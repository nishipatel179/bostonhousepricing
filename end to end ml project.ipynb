```python
import pandas as pd
import numpy as np

# Download the raw dataset from the original source
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# Process the data
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# Convert to DataFrame for easier handling
boston_df = pd.DataFrame(data, columns=[
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"
])
boston_df['MEDV'] = target

# Check the first few rows of the DataFrame
print(boston_df.head())
```

```python
# independent and dependent features

x=boston_df.iloc[:,:-1] # including all features and leaving the target variable
y=boston_df.iloc[:,-1]
```

```python
x.head()
```

```python
#train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
#standard scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
# print coefficient and intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
##on which parametes the model has been trained
regression.get_params()
```

```python
## prediction with test data
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

```python
## plot a scatter plot for prediction
plt.scatter(reg_pred,y_test)
```

```python
## residual means error
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## plot this residuals
import seaborn as sns
sns.displot(residuals,kind="kde")
```

```python
## scatter plot with result to prediction and residuals
#uniform distribution is there
plt.scatter(reg_pred,residuals)
## didnot follwoing any pattern
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Example usage
print(mean_squared_error(reg_pred, y_test))
print(mean_absolute_error(reg_pred, y_test))
print(np.sqrt(mean_squared_error(reg_pred,y_test)))
```

```python
# r square and adjusted r square
from sklearn.metrics import  r2_score
score=r2_score(y_test,reg_pred)
print(score)
```
